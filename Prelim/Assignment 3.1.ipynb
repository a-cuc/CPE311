I approached the problem by doing it by phrases/words.

```
You are given an integer n. Determine if there is an unconnected graph with n vertices that contains at
least two connected components and contains the number of edges that is equal to the number of vertices.
Each vertex must follow one of these conditions:

Its degree is less than or equal to 1.
It's a cut-vertex

The graph must be simple.
Loops and multiple edges are not allowed.
Input format:

First line: n
Output format: Print Yes if it is an unconnected graph. Otherwise, print No.
Constraints: 1 <= n <= 100

```
The first constraint is that there should be at least <font color = "blue"> **two** </font> <font color = "red"> connected components</font>, which means that any ```n``` less than 4 should already result to a **No**.

The next constraint is that the components should contain the same number of edges and vertices.

A graph that satisfies that constraint is a triangle, with 3 sides and 3 edges.
Considering that there should at least be two of them, any ```n``` less than 6 should also result to a **No**.

The next constraint is that <font color = "blue"> each vertex should either have at most 1 degree </font> **OR** <font color = "red"> it is a cut vertex </font>. A triangle graph alone is not a cut vertex since if you cut a triangle, it would not split into two.

And we can conclude that any ```n``` less than 12 would automatically result to a **No**.

````python
# Implementing this in code is a simple if else function
def unconnectedGraphCheck(n):
  if 1 <= n and n <= 100: # constraint of 1 <= n <= 100
    if n >= 12:
      print("Yes")
    else:
      print("No")


unconnectedGraphCheck(10)
unconnectedGraphCheck(12)
unconnectedGraphCheck(14)
unconnectedGraphCheck(101)
unconnectedGraphCheck(-10)
````

However, since the assignment involves making a graph, I decided to make one that satisfies the constraints

````python
def createGraph(n):
  base = {
      1 : {2,3,4},
      2 : {3,5},
      3 : {6},
      7 : {8,9,10},
      8 : {9,11},
      9 : {12}
  }
  if n >= 12: # returns the base graph if n = 12
    for i in range(n-12): # appends a vertex to the last number vertex
      base[12+i] = {13+i}
    return base

result_graph = createGraph(15)
print(createGraph(15))
````
