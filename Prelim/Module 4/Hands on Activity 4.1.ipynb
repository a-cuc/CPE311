The problem I am trying to solve is the <font color = "blue">Tower of Hanoi</font> problem.

The solution I made assumes the following:
*   The user input is n where n is the number of blocks.
*   There are 3 "pegs" for the blocks to be able to be placed.

The solution below follows the algorithm in the Hands-On Activity

```python
def solve(n, peg1 = "source", peg2 = "auxiliary", peg3 ="destination"): # recursive

  if n == 0: # base cases
    print("No n!")
  if n == 1:
    print("Moved disk #" + str(n) + " from " + peg1 + " to " + peg3)

  else:
    solve(n-1, peg1, peg3, peg2) # moves all disks to auxiliary until 1 is left
    print("Moved disk #" + str(n) + " from " + peg1 + " to " + peg3)
    solve(n-1, peg2, peg1, peg3)

toSolve = int(input("Enter n: "))
solve(toSolve)
```

To understand the solution, I have added counters of n before the first and second recursions, added a print function to see the current pegs, and made the peg names be A, B, and C.

```python
def solve(n, peg1 = "A", peg2 = "B", peg3 ="C"): # recursive
  print("peg1 = " + peg1 + ", peg2 = " + peg2 + ", peg3 = " + peg3)
  if n == 0: # base cases
    print("No n!")
  if n == 1:
    print("Moved disk #" + str(n) + " from " + peg1 + " to " + peg3)

  else:
    print("n = " + str(n) + " before first recur")
    solve(n-1, peg1, peg3, peg2) # moves all disks to auxiliary until 1 is left
    print("Moved disk #" + str(n) + " from " + peg1 + " to " + peg3)
    print("n = " + str(n) + " before second recur")
    solve(n-1, peg2, peg1, peg3)

solve(3)
```

The second recursion is then executed while n-1, which means that the pegs are in position **A,C,B**. The main function then executes which transfers the biggest block to the destination peg.
