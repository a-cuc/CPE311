Pandas Dataframe Creation and Methods

```python
# Cycling Distance
cycling_data = [10,7,0,None,2,4,15,3,10,9,0,None]

joined_data = list(zip(step_data, cycling_data)) # joins cycling and step data

joined_data
```

```python
# Create Dataframe
activity_df = pd.DataFrame(joined_data)
activity_df
```

```python
# Adds column names
activity_df = pd.DataFrame(joined_data, index=pd.date_range('20240306', periods=5), columns = ('walking', 'cycling'))
activity_df
```

Indexing with Dataframes

Can be indexed using:
* loc - > named index value
* iloc - > integer index

```python
activity_df.loc['2024-03-06']
```

```python
activity_df.iloc[0]
```

```python
activity_df.loc[:'2024-03-09'] # you can use .loc on dataframes for slicing
```

```python
activity_df.iloc[:2] # you can use .iloc on dataframes for slicing
```

```python
activity_df.iloc[:, :] # row, column
```

```python
filepath
df = pd.read_csv(filepath, index_col = 'Id') # uses index_col to set index column from file
df
```

```python
df.head(5) # views the first n data
```

```python
df.tail() # views last data
```

```python
df['SepalArea'] = df.SepalLengthCm * df.SepalWidthCm
df
```

```python
# Add new column for the petal area
df['PetalArea'] = df.PetalLengthCm * df.PetalWidthCm
df
```

```python
df['abbrev'] = (df.Species.apply)(lambda x:x.replace('Iris-', ''))
# cleans data by removing the Iris- in Species column
df
```

```python
df.drop(['Species'], axis = 1, inplace = True) # removes the Species column
# axis specifies whether row or column
df
```

```python
small_data = pd.concat([df.iloc[:2], df.iloc[-2:]]) # gets the first and last 2 data and joins
small_data
```

```python
group_size = df.groupby('abbrev').size()
group_size
```
