<h2>Exercise 1</h2>

Run the given code below for exercises 1 and 2, perform the given tasks without using any Python modules.


```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the
standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are
obtained when using the statistics module (where possible):

* Mean

* Median

* Mode (hint: check out the Counter in the collections module of the standard library at
https://docs.python.org/3/library/collections.html#collections.Counter)

* Sample variance

* Sample standard deviation

```python
def mean(x):
  return sum(x)/len(x)

mean(salaries)
```

```python
# using Statistics module
import statistics
print(statistics.mean(salaries))
```

```python
def median(x):
  x = sorted(x)
  index = 0
  if len(x) % 2 == 0: # even
    index = int(len(x) / 2)
    return (x[index] + x[index-1])/2
  else: # odd
    index = int((len(x)/2) + 1)
    return x[index-1]


median(salaries)
```

```python
# using Statistics module
import statistics
print(statistics.median(salaries))
```

```python
import collections

def mode(x):
  x = collections.Counter(x).most_common()
  return x[0][0]
mode(salaries)
```

```python
# using Statistics module
import statistics
print(statistics.mode(salaries))
```

```python
def sample_variance(x):
    n = len(x)
    m = mean(x)
    table = [(element - m) ** 2 for element in x]
    return sum(table) / (n - 1)

sample_variance(salaries)
```

```python
# using Statistics module
import statistics
print(statistics.variance(salaries))
```

```python
def sample_stdev(x):
  return sample_variance(x)**0.5

sample_stdev(salaries)
```

```python
# using Statistics module
import statistics
print(statistics.stdev(salaries))
```

<h2>Exercise 2</h2>

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:

* Range

* Coefficient of variation

* Interquartile range

* Quartile coefficient of dispersion

```python
def range_fn(x): # does not use the statistics module
  x = sorted(x)
  return x[-1] - x[0]

range_fn(salaries)
```

```python
import statistics

def variation_coefficient(x):
  return statistics.stdev(x) / statistics.mean(x)

variation_coefficient(salaries)
```

```python
import statistics

def iqr(x):
    q1, q2, q3 = statistics.quantiles(x)

    return q3 - q1

iqr(salaries)
```

```python
import statistics

def qcd(x):
  q1, q2, q3 = statistics.quantiles(x)

  return (q3-q1)/(q3+q1)

qcd(salaries)
```
