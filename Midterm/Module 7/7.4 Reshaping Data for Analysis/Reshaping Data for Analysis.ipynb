```python
import pandas as pd
long_df = pd.read_csv(
  'data/long_data.csv',
  usecols=['date', 'datatype', 'value'] # only gets date, datatype, and value columns
).rename(
  columns={
    'value' : 'temp_C' # renames value to temp_C
  }
).assign(
  date=lambda x: pd.to_datetime(x.date), # converts date to date dtype
  temp_F=lambda x: (x.temp_C * 9/5) + 32 # adds farenheit column
)
long_df.head()
```

```python
long_df.head().T # transposes row and column
```

```python
pivoted_df = long_df.pivot(
  index='date', columns='datatype', values='temp_C'
) # index contains the index, columns for the column names
pivoted_df.head()
```

```python
pd.pivot( # same result
  index=long_df.date.name, columns=long_df.datatype.name, values=long_df.temp_C.name
          , data = long_df # added .name and data args for the code to run
).head()
```

```python
pivoted_df.describe()
```

```python
pivoted_df = long_df.pivot(
  index='date', columns='datatype', values=['temp_C', 'temp_F'] # multiple values
)
pivoted_df.head()
```

```python
pivoted_df['temp_F']['TMIN'].head()
```

```python
multi_index_df = long_df.set_index(['date', 'datatype']) # multi indexing allows for multiple index values
multi_index_df.index
```

```python
multi_index_df.head()
```

```python
unstacked_df = multi_index_df.unstack()
unstacked_df.head()
```

```python
extra_data = long_df.append(
  [{'datatype' : 'TAVG', 'date': '2018-10-01', 'temp_C': 10, 'temp_F': 50}]
).set_index(['date', 'datatype']).sort_index()
extra_data.head(8)
```

```python
extra_data.unstack().head()
```

```python
extra_data.unstack(fill_value=-40).head()
# using fill_value only fills with a fixed value
```

```python
wide_df = pd.read_csv('data/wide_data.csv')
wide_df.head() # converting from wide data to long data
```

```python
melted_df = wide_df.melt(
  id_vars='date',
  value_vars=['TMAX', 'TMIN', 'TOBS'],
  value_name='temp_C',
  var_name='measurement'
)
melted_df.head()
```

```python
pd.melt(
  wide_df,
  id_vars='date',
  value_vars=['TMAX', 'TMIN', 'TOBS'],
  value_name='temp_C',
  var_name='measurement'
).head()
```

```python
wide_df.set_index('date', inplace=True)
wide_df.head()
```

```python
stacked_series = wide_df.stack()
stacked_series.head()
```

```python
stacked_df = stacked_series.to_frame('values')
stacked_df.head()
```

```python
stacked_df.index
```

```python
stacked_df.index.names
```

```python
stacked_df.index.rename(['date', 'datatype'], inplace=True)
stacked_df.index.names
```
