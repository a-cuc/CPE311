```python
filepath = '/content/spambase.csv'
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv(filepath)

data
```

```python
data.describe() # descriptive data analysis for entire dataset
```

```python
# inserts a new row Classification which display spam if value of class is 1, else not spam
data['Classification'] = (data['class'].apply)(lambda x : 'Spam' if x == 1 else 'Not Spam')

data
```

```python
spam = data[data.Classification == 'Spam'] # gets email labelled as spam

spam
```

```python
spam.describe() # descriptive data analysis for spam dataset
```

```python
# gets the mean for the specified word frequency
spam_word_freq_mean = np.mean(spam.iloc[:, :48])

print(np.mean(spam_word_freq_mean))
```

```python
# gets the mean for the specified character frequency
spam_char_freq_mean = np.mean(spam.iloc[:, 48:54])

print(np.mean(spam_char_freq_mean))
```

```python
# gets the mean uninterrupted sequences of capital letters
spam_capital_run_mean = np.mean(spam.iloc[:, 56])

print(spam_capital_run_mean)
```

```python
notSpam = data[data.Classification == 'Not Spam'] # gets email labelled as not spam

notSpam
```

```python
notSpam.describe() # descriptive data analysis for non-spam dataset
```

```python
# gets the mean for the specified word frequency
notSpam_word_freq_mean = np.mean(notSpam.iloc[:, :48])

print(np.mean(notSpam_word_freq_mean))
```

```python
# gets the mean for the specified character frequency
notSpam_char_freq_mean = np.mean(notSpam.iloc[:, 48:54])

print(np.mean(notSpam_char_freq_mean))
```

```python
# gets the mean uninterrupted sequences of capital letters
notSpam_capital_run_mean = np.mean(notSpam.iloc[:, 56])

print(notSpam_capital_run_mean)
```
